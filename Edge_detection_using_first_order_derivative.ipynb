```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/ganesha.jpg")
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
x_sobel = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
```

```python
plt.imshow(x_sobel, cmap='gray')
```

```python
y_sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 5)
```

```python
plt.imshow(y_sobel, cmap='gray')
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
plt.imshow(laplacian, cmap = 'gray')
```

```python
threshold_1, threshold_2 = 120, 200
```

```python
canny_ = cv2.Canny(img, threshold_1, threshold_2)
```

```python
plt.imshow(canny_, cmap='gray')
```
